```python
from qiskit import*
from math import pi
```

```python
circuit = QuantumCircuit(3,3)

circuit.rz(pi/2,2)
circuit.sx(2)
circuit.rz(pi/2,2)
#circuit.barrier()
circuit.cx(1,2)
circuit.rz(-pi/4,2)
circuit.cx(0,2)
circuit.rz(pi/4,2)
circuit.cx(1,2)

circuit.rz(-pi/4,2)
circuit.cx(0,2)

circuit.rz(pi/4,1)
circuit.cx(0,1)
circuit.rz(pi/4,0)
circuit.rz(-pi/4,1)
circuit.cx(0,1)
#circuit.barrier()
circuit.rz((3*pi)/4,2)
circuit.sx(2)
circuit.rz(pi/2,2)
```

```python
%matplotlib inline
circuit.draw(output="mpl")
```

```python
import qiskit.quantum_info as qi

op = qi.Operator(circuit)
print(op)
```

## The resulting matrix is similar to the matrix of a Tofolli gate with some phase change. Hence, it has been successfully decomposed.

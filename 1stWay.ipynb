```python
from qiskit import*
```

```python
qc = QuantumCircuit(3,3)
```

```python
%matplotlib inline
```

```python
qc.draw(output="mpl")
```

```python
qc.h(2)
qc.cx(1,2)
qc.tdg(2)
qc.cx(0,2)
qc.t(2)
qc.cx(1,2)
qc.tdg(2)
```

```python
qc.draw(output="mpl")
```

```python
qc.cx(0,2)
```

```python
qc.draw(output="mpl")
```

```python
qc.barrier()
```

```python
qc.draw(output="mpl")
```

```python
qc.t(1)
qc.t(2)
```

```python
qc.draw(output="mpl")
```

```python
qc.cx(0,1)
qc.h(2)
```

```python
qc.draw(output="mpl")
```

```python
qc.t(0)
qc.tdg(1)
qc.cx(0,1)
qc.draw(output="mpl")
```

```python
import qiskit.quantum_info as qi

op = qi.Operator(qc)
print(op)
```

## The resulting matrix is similar to the matrix of a Tofolli gate. Hence, it has been successfully decomposed.
